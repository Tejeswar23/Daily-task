```python
# student Academic  Performance
import random
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
n = 200
student_id = np.arange(1,n+1)
maths = np.random.randint(35,100,n)
science = np.random.randint(35,100,n)
english = np.random.randint(35,100,n)
attendence_percent = np.random.randint(60,100,n)
data = {
    'student_id':student_id,
    'maths': maths,
    'science':science,
    'english':english,
    'attendence_percent':attendence_percent
    
}
df = pd.DataFrame(data)
print(df)
```

```python
#Compute total and average scores.
maths_total = df["maths"].sum()
science_total =df['science'].sum()
english_total = df["english"].sum()
print(f"total marks : {maths_total+science_total+english_total}")
maths_percentage = (maths_total/20000)*100
science_percentage = (science_total/20000)*100
english_percentage = (english_total/20000)*100
print("Maths percentage :",maths_percentage)
print("Science percentage :",science_percentage)
print("English percentage :",english_percentage)
# classify students based on attendence percentage
top_Students = df.loc[df['attendence_percent']> 60]
print(top_Students)
```

```python
result = df[(df['maths'] < 50) & (df['science']< 50)& (df['english'])&(df['attendence_percent'])]
print(result)
result_1 = df.groupby(df['english'],df['science'],df['maths']).s
```

```python
# Standardize subject scores.& assigning grades
eng_per = english_percentage
sci_per = science_percentage
mat_per = maths_percentage
if eng_per > 80 and (science_percentage and mat_per):
    print("A grade")
elif (eng_per > 70 and eng_per < 80) and (sci_per > 70 and sci_per < 80) and (mat_per > 70 and mat_per <80):
    print("B grade")
else:
    print("C Grade")
```

```python
# visualization
# Pie chart
# Pass Percentage subject wise
names = ['english','maths','science']
values = [maths_percentage,science_percentage,english_percentage]
plt.pie(values,labels=names,autopct="%1.1f%%",shadow=True)
plt.title("Pass Percentage subject wise")
plt.show()
```
