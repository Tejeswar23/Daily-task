```python
import numpy as np
import pandas as pd
import random
marks = np.random.randint(0, 101, size=(200, 6))
#total per student
total_marks = marks.sum(axis=1)
percentage = (total_marks / 600) * 100
# print(total_marks)
# Grades assigning
grades = np.where(percentage >= 90, 'A',
            np.where(percentage >= 75, 'B',
            np.where(percentage >= 60, 'C',
            np.where(percentage >= 40, 'D', 'F'))))
# top 10 students
top_10_cutoff = np.percentile(percentage, 90)
top_10_students = percentage >= top_10_cutoff
print(top_10_cutoff)


# 2 - Pandas
data = {
    "student_id": np.repeat(np.arange(1, 201), 6),
    "subject": ["Math", "Physics", "Chemistry", "Biology", "English", "CS"] * 200,
    "marks": marks.flatten(),
    "exam_date": pd.to_datetime("2025-01-10")
}
df = pd.DataFrame(data)
print(df)
# failed students
df["pass"] = df["marks"] >= 40
subject_pass_percentage = (df.groupby("subject")["pass"].mean() * 100)
print(subject_pass_percentage)
# Student fail in more then one subject
fail_counts = (
    df[df["marks"] < 40].groupby("student_id").size())
students_failing_multiple = fail_counts[fail_counts > 1]
print(students_failing_multiple)

```
