```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
df = pd.read_csv('StudentPerformance.csv')
print(df.head())
print(df.shape)
print(df.info())
print(df.describe())
```

```python
# hours studied vs performance Index by using bar graph
x = df['Hours Studied']
y = df['Performance Index']
plt.bar(x,y)
plt.show()
```

```python
x = df['Previous Scores']
y = df['Performance Index']
plt.bar(x,y)
plt.xlabel('Previous Scores')
plt.ylabel('Performance Index')
plt.show()
```

```python
activity_counts = df['Extracurricular Activities'].value_counts()
plt.pie(
    activity_counts,
    labels=activity_counts.index,autopct='%1.1f%%',startangle=90)
plt.title('Extracurricular Activities Distribution')
plt.show()
```
