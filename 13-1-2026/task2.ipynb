```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
df = pd.read_csv('StudentPerformance.csv')
print(df.shape)
```

```python
# Using the info method to know more about the data set.
df.info()
```

```python
# Using the describe method to know the mean and mode like values.
df.describe()
```

```python
# Average time spent on studies.
avg_time = df['Hours Studied'].mean()
print("Average time spent on studies ",avg_time)
```

```python
# Previous Score percentage and Current Score percentage.
prev_score = df['Previous Scores'].mean()
print(f"Previous score pass percentage:{prev_score}")
curr_score = df['Performance Index'].mean()
print(f"Current Score pass percentage :{curr_score}")
```

```python
# More study time increase performance index.
x = df['Hours Studied']
y = df['Performance Index']
plt.scatter(x,y)
plt.xlabel("Hours Studied")
plt.ylabel("Performance Index")
plt.title("Study Time vs Performance Index")
plt.show()
```

```python
# Average performance index for different study hour groups.
avg_perf = df.groupby("Hours Studied")["Performance Index"].mean()
print(avg_perf)
plt.bar(avg_perf.index, avg_perf.values)
plt.xlabel("Hours Studied")
plt.ylabel("Average Performance Index")
plt.title("Average Performance Index by Study Hours")
plt.show()
```

```python
# Students participate in extra circular activities.
activity_counts = df['Extracurricular Activities'].value_counts()
plt.pie(
    activity_counts,
    labels=activity_counts.index,autopct='%1.1f%%',startangle=90)
plt.title('Extracurricular Activities Distribution')
plt.show()
```

```python
# Students distributed based on their performance index scores.
# bins - split the data into groups.
plt.hist(df['Performance Index'], bins=10)
plt.xlabel('Performance Index')
plt.ylabel('Frequency')
plt.title('Performance Index Distribution')
plt.show()
```

```python
# Average scores of students sleeping more than 7 hours vs less than 7 hours.
more_sleep = df[df["Sleep Hours"] > 7]["Performance Index"].mean()
less_sleep = df[df["Sleep Hours"] < 7]["Performance Index"].mean()
print("Average Performance 'Sleep > 7 hrs':", more_sleep)
print("Average Performance 'Sleep < 7 hrs':", less_sleep)
```

```python
# Performance based on number of sample papers practiced.
x = df['Sample Question Papers Practiced']
y = df['Performance Index']
plt.bar(x,y)
plt.xlabel("Number of Sample Papers Practiced")
plt.ylabel("Average Performance Index")
plt.title("Performance vs Sample Papers Practiced")
plt.show()
```
